```python
#Data Analysis Project  - Store Profit and Sales Analysis
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default="plotly_white"
file_path="../Dataset/SuperStore_Sales_Dataset.csv"
data =pd.read_csv(file_path,encoding="latin-1")
print(data.head())
print(data.describe())

data['Order Date']=pd.to_datetime(data['Order Date'],format="%d-%m-%Y")
data['Ship Date']= pd.to_datetime(data['Ship Date'],format="%d-%m-%Y")
# covert month year and days of week
data['Order Month']= data['Order Date'].dt.month
data['Order Year']=data['Order Date'].dt.year
data['Order Day of week']=data['Order Date'].dt.day_of_week


#lets see monthly Sales
sales_by_month=data.groupby('Order Month')['Sales'].sum().reset_index()
#print(data.head())
fig = px.line(sales_by_month, x='Order Month', y='Sales', title='Monthly Sales Analysis')
fig.show()

# sales by category 
sales_by_cat= data.groupby('Category')['Sales'].sum().reset_index()
fig =px.pie(sales_by_cat, values="Sales", names= "Category", hole=0.5, color_discrete_sequence=px.colors.qualitative.Pastel, title='Monthly Sales Analysis')
fig.update_traces(textposition = 'inside' , textinfo ='percent+label' )
fig.update_layout(title="Sales Analysis By Cat",title_font=dict(size=40))
fig.show()

# Sales by sub Category
sales_by_sub_cat=data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_sub_cat, x='Sub-Category' , y='Sales', title='Monthly Sales Analysis')
fig.show()

#Profile Analysis
profit_by_month=data.groupby('Order Month')['Profit'].sum().reset_index()
fig =px.line(profit_by_month, x='Order Month', y='Profit', title='Profit By Month')
fig.show()

#Sales and profit analysis by customer segement
sales_profit_by_segemnt=data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
color_palette =colors.qualitative.Pastel
fig =go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segemnt['Segment'],y=sales_profit_by_segemnt['Sales'],name='Sales',marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=sales_profit_by_segemnt['Segment'],y=sales_profit_by_segemnt['Profit'],name='Profit',marker_color=color_palette[1]))
fig.update_layout(title='Profit and Sales analysis by customer segment', xaxis_title='Customer Segment',yaxis_title='Amount')
fig.show()

sales_profit_by_segemnt= data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segemnt['Sales_to_profit_Ratio'] = sales_profit_by_segemnt['Sales']/sales_profit_by_segemnt['Profit']
print(sales_profit_by_segemnt[['Segment', 'Sales_to_profit_Ratio']])
```
