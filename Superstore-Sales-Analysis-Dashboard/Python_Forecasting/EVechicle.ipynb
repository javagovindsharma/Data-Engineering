```python
import os
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "browser"

file_path="../../Dataset/Electric_Vehicle_Population_Data.csv"
#print("File Exists:", os.path.exists(file_path))

data = pd.read_csv(file_path)
#print(data.head())
data['Country']=data['Country'].fillna("NA")
data['City']=data['City'].fillna("NA")
data['Postal Code']=data['Postal Code'].fillna(0)
data['Legislative District']=data['Legislative District'].fillna(0)
data['Vehicle Location']=data['Vehicle Location'].fillna("POINT (0)")
data['Electric Utility']=data['Electric Utility'].fillna("NA")
data['2020 Census Tract']=data['2020 Census Tract'].fillna(0)

#country_data=data.groupby("Model Year")['Country'].sum().reset_index()
#print(country_data.count())
bins = np.arange(1990, data['Model Year'].max() + 6, 6)
data['Year_Group'] = pd.cut(
    data['Model Year'],
    bins=bins,
    labels=[f"{i}-{i+5}" for i in range(1990, data['Model Year'].max(), 6)]
)
result = data.groupby('Year_Group', observed=True)['Legislative District'].sum().reset_index()
print(result['Legislative District'].values)

#print(data.isnull().sum())
fig=  px.pie(
    result,
    names='Year_Group',                 # category
    values='Legislative District',      # numeric
    title="Based on Year Legislative District Data",
    hole=0.5
)

fig.show()
```
